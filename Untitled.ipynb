```python
import numpy as np
import cv2
import os;
import matplotlib.pyplot as plt
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt2.xml')
directory = sorted(os.listdir("../../../Downloads/thumbnails_features_deduped_publish (1)/thumbnails_features_deduped_publish/"));
for label in directory:    
    images = os.listdir("../../../Downloads/thumbnails_features_deduped_publish (1)/thumbnails_features_deduped_publish/"+label+"/");
    APPROX = 0;
    print len(images)
    os.mkdir('vigneshwaran');
    for i in images:
        count = 0;
        img = cv2.imread('../../../Downloads/thumbnails_features_deduped_publish (1)/thumbnails_features_deduped_publish/'+label+'/'+i,0);
        gray = img;
        faces = face_cascade.detectMultiScale(gray,1.05,5)
        for (x,y,w,h) in faces:
            count = count+1;
            roi_gray = gray[y-APPROX:y+h+APPROX, x-APPROX:x+w+APPROX]
            roi_gray = cv2.resize(roi_gray, (100,100))
            cv2.imwrite('vigneshwaran/'+str(i),roi_gray)
            roi_gray=cv2.flip(roi_gray,1)
            cv2.imwrite('vigneshwaran/'+'mirror'+str(i),roi_gray)
            break;
```
